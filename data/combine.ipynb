```python
%pip install plotly
import plotly.express as px
import pandas as pd
import os

data_0 = pd.read_csv('daily_sales_data_0.csv')
data_1 = pd.read_csv('daily_sales_data_1.csv')
data_2 = pd.read_csv('daily_sales_data_2.csv')

new_data = pd.concat(
    [data_0[data_0['product'] == 'pink morsel'], data_1[data_1['product'] == 'pink morsel'], data_2[data_2['product'] == 'pink morsel']],
    ignore_index=True
    )


new_data['price'] = new_data['price'].map(lambda x: float(x[1:]))
new_data['sales'] = new_data['price'] * new_data['quantity']
new_data
```

```python
output_data = new_data[['sales','date', 'region']]
output_data['date'] = pd.to_datetime(output_data['date'])
output_data.sort_values(by=['date'])
parent = os.path.dirname(os.getcwd())
output_data.to_csv(f'{parent}/output.csv', index=False)
```

```python
fig = px.line(output_data, x='date', y='sales', title='Total Sales of Pink Morsels by Date')
fig.show()
```
